<h3>Matplot Lib</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [0, 1, 2]
y = [100, 200, 300]
```

```python
plt.plot(x, y)
```

```python
housing = pd.DataFrame({
    'rooms': [1, 1, 2, 2, 2, 3, 3, 3],
    'price': [100, 120, 190, 200, 130, 310, 330, 305]
})
housing
```

```python
plt.scatter(housing.rooms, housing.price)
```

```python
plt.plot(x, y, color='red', marker='o', markersize=15)
plt.title('Title')
plt.xlim(0, 2)
plt.ylim(0, 300)
plt.xlabel('X Label')
plt.ylabel('Y Label')
```
