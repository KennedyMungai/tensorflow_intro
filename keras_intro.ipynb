<h1>Keras Intro</h1>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
df = pd.read_csv('./data_sets/fake_reg.csv')
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1', 'feature2']].values
X
```

```python
y = df['price'].values
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train, X_test
```

```python
X_train.max(), X_test.max()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(4 , activation='relu'))
model.add(Dense(4 , activation='relu'))
model.add(Dense(4 , activation='relu'))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mse')
```

```python
model.fit(x=X_train, y=y_train, epochs=250, verbose=1)
```

```python
model.history.history
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df
```

```python
loss_df.plot()
```

```python
model.evaluate(X_test, y_test, verbose=0)
```

```python
model.evaluate(X_train, y_train, verbose=0)
```

```python
test_predictions = model.predict(X_test)
```

```python
len(test_predictions)
```

```python
test_predictions = pd.Series(test_predictions.reshape(300,))
```

```python
test_predictions
```

```python
pred_df = pd.DataFrame(y_test, columns=['Test True Y'])
pred_df
```

```python
pred_df = pd.concat([pred_df, test_predictions], axis=1)
```

```python
pred_df
```

```python
pred_df.columns = ['Test True Y', 'Model Predictions']
pred_df
```

```python
sns.scatterplot(x='Test True Y', y='Model Predictions', data=pred_df)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(pred_df['Test True Y'], pred_df['Model Predictions'])
```

```python
mean_squared_error(pred_df['Test True Y'], pred_df['Model Predictions'])
```

```python
# Root mean squared error
np.sqrt(mean_squared_error(pred_df['Test True Y'], pred_df['Model Predictions']))
```

```python
X = df.drop('price', axis=1).values
```

```python
y = df.price.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
X_train.shape
```

```python
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

model.add(Dense(1, activation='relu'))

model.compile(loss='mse', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), batch_size=128, epochs=400, verbose=1)
```

```python
model.history.history
```

```python
losses = pd.DataFrame(model.history.history)
losses.head()
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_squared_error(y_test, predictions)
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
mean_absolute_error(y_test, predictions)
```
